```python
df.Gender = np.where(df.Gender.str.contains('Transgender', case=False),'Transgender',df['Gender'])
df.Gender = np.where(df.Gender.str.contains('Intersex', case=False),'Transgender',df['Gender'])
df.Gender = np.where(df.Gender.str.contains('Eunuch', case=False),'Transgender',df['Gender'])
df.Gender = np.where(df.Gender.str.contains('Binary', case=False),'Transgender',df['Gender'])
df.Gender = np.where(df.Gender.str.contains('Male',),'Male',df['Gender'])
df.Gender = np.where(df.Gender.str.contains('Female', case=False),'Female',df['Gender'])

df.Gender.value_counts(dropna=False)
```

```python
plt.figure(figsize=(8,6))
sns.countplot(y='Occupation', data=df, hue='Gender',
             order=df.Occupation.value_counts().sort_values(ascending=False)[:10].index);
```

```python
df['century'] = df['Death year'] // 100 + 1
df['century'].value_counts()
```
