## how many employees working for each department?

```python
plt.figure(figsize=(12,6))
ax=sns.countplot(x='department_name', data=df,
            order=df.department_name.value_counts(ascending=False).index)
plt.xticks(rotation=45)

labels = df.department_name.value_counts(ascending=False).values

ax.bar_label(container=ax.containers[0], labels=labels);
```

## how many employees working with different job title?

```python
plt.figure(figsize=(12,6))
ax=sns.countplot(x='job_title', data=df,
            order=df.job_title.value_counts(ascending=False).index)
plt.xticks(rotation=90)

labels = df.job_title.value_counts(ascending=False).values

ax.bar_label(container=ax.containers[0], labels=labels);
```
