## which Category has highest sale

```python
pd.pivot_table(store, values='Sales',index=['Category'], aggfunc=sum).sort_values(by='Sales', ascending=False).round(0)
```

```python
def value_label(x_list, y_list):
    for i in range(len(x_list)):
        plt.annotate(y_list[i],(i,y_list[i]), ha='center', )
        
sns.barplot(x='Category', y='Sales', data=store, estimator=sum, ci=None,
            order=store.groupby('Category').Sales.sum().sort_values(ascending=False).index);

index=store.groupby('Category').Sales.sum().sort_values(ascending=False).index
values=store.groupby('Category').Sales.sum().sort_values(ascending=False).values.round(0)

value_label(index, values)
```
