## Dealing with null values:

```python
# Linear feet of street connected to property, values are in float and replace the missing values with mean
house.LotFrontage.fillna(house.LotFrontage.mean(), inplace=True)
```

```python
# Masonry veneer area in square feet, if there is not masonry then  area will be 0
house.MasVnrArea.fillna((0), inplace=True)
```

```python
# if values is missing then replace with most occuring electrical system ie. mode()
house.Electrical.fillna(house.Electrical.mode()[0], inplace=True)
```

```python
# this is garage built year, if missing then fill it with most occuring year value using mode()
house.GarageYrBlt.fillna(house.GarageYrBlt.mode()[0], inplace=True)
```

```python
# if there is no alley available then fill nan with No Allay
house.Alley.fillna('No Alley', inplace=True) 
```

```python
# if the fireplace quality is missing then indiacte that as No Fireplace
house.FireplaceQu.fillna('No Fireplace', inplace=True)
```

```python
# Masonry veneer type, if missing the fill with "No Masonry"
house.MasVnrType.fillna('No Masonry', inplace=True)
```

```python
# this column indicates quality of the basement according to height, if missing then we can say the basement is missing
house.BsmtQual.fillna('No Basement', inplace=True)
```

```python
# this column indicates the condition of basement, if missing then there is no basement.
house.BsmtCond.fillna('No Basement', inplace=True)
```

```python
# this column indicates basement explosure, if missing then there is no basement.
house.BsmtExposure.fillna('No Basement', inplace=True)
```

```python
# this column indicates rating of basement finished area, if missing then we can say that there is no basement
house.BsmtFinType1.fillna('No Basement', inplace=True)
house.BsmtFinType2.fillna('No Basement', inplace=True)
```

```python
# these columns refered to Garage, if missing then we can fill the missing values with "No Garage"
house.GarageType.fillna('No Garage', inplace=True)
house.GarageFinish.fillna('No Garage', inplace=True)
house.GarageQual.fillna('No Garage', inplace=True)
house.GarageCond.fillna('No Garage', inplace=True) 
```

```python
# this column indicates the pool quality, if missing then fill with "No Pool"
house.PoolQC.fillna('No Pool', inplace=True)
```

```python
# this column indicates the fence quality, if missing then we can assume that fence is absent
house.Fence.fillna('No Fence', inplace=True)
```

```python
# if there is no any miscellenoues feature then fill missing values with "No Miscellaneous"
house.MiscFeature.fillna('No Miscellaneous', inplace=True)
```

## House prices with respect to house Style

```python
pd.pivot_table(house, index='HouseStyle', values='SalePrice', aggfunc='mean').sort_values(by='SalePrice', ascending=False).round(2)
```

```python
order=house.groupby('HouseStyle').SalePrice.mean().sort_values(ascending=False).index
```

```python
def add_value_label(x_list, y_list):
    for i in range(len(x_list)):
        plt.annotate(y_list[i], (i,y_list[i]), ha='center')

plt.figure(figsize=(10,5))
sns.barplot(x='HouseStyle', y='SalePrice', data=house, ci=0, order=order);

index=house.groupby('HouseStyle').SalePrice.mean().sort_values(ascending=False).index
values=house.groupby('HouseStyle').SalePrice.mean().sort_values(ascending=False).round(2).values

add_value_label(index,values);
```

#### inference:
1. 2.5Fin style house has most average selling price
2. 1.5Unf style house has least average selling price
