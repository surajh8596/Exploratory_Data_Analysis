# Analyzation of data

## Check who is the highest run scorer in all IPL

```python
pd.pivot_table(df, index='striker', values='runs_off_bat', aggfunc=sum).sort_values(by='runs_off_bat', ascending=False)[:10]
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='striker', y='runs_off_bat', data=df, estimator=sum, ci=None,
           order=df.groupby('striker').runs_off_bat.sum().sort_values(ascending=False)[:10].index)
plt.xlabel('Player', fontsize=15)
plt.ylabel('Total runs', fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

# Analysis on IPL 2008 season

```python
IPL_2008=df[df['year']==2008]
IPL_2008
```

## Which player scored most runs in IPL 2008

```python
pd.pivot_table(IPL_2008, index='striker', values='runs_off_bat', aggfunc=sum).sort_values(by='runs_off_bat', ascending=False)[:10]
```

```python
def add_value_label(x_list, y_list):
    for i in range(len(x_list)):
        plt.annotate(y_list[i], (i,y_list[i]), ha='center')
        
plt.figure(figsize=(10,6))
sns.barplot(x='striker', y='runs_off_bat', data=IPL_2008, estimator=sum, ci=None,
           order=IPL_2008.groupby('striker').runs_off_bat.sum().sort_values(ascending=False)[:10].index)
plt.xticks(rotation=90)
plt.xlabel('Player', fontsize=15)
plt.ylabel('Total Runs Scored', fontsize=15)

index=IPL_2008.groupby('striker').runs_off_bat.sum().sort_values(ascending=False)[:10].index
value=IPL_2008.groupby('striker').runs_off_bat.sum().sort_values(ascending=False)[:10].values

add_value_label(index, value);
```

## What is the most common wicket type in IPL 2008

```python
ax=sns.countplot(y='wicket_type', data=IPL_2008,
                order=IPL_2008.wicket_type.value_counts(ascending=False).index)
value=IPL_2008.wicket_type.value_counts(ascending=False).values
ax.bar_label(container=ax.containers[0], labels=value);
```
